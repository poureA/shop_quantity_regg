```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
file_train = pd.read_csv('F:\\datasets\\shop_regg\\train.csv')
df_train = file_train.copy()
df_train.sample(10)
```

```python
df = df_train.dropna()
df.info()
```

```python
label = df['quantity']
df = df.drop(['quantity','id','date','shop'],axis=1)
df_num = df.drop(['city','brand','container','capacity'],axis=1)
df_nan = df.drop(df_num.columns,axis=1)
```

```python
encoded_nan = pd.get_dummies(df_nan)
```

```python
concated_df = pd.concat([df_num,encoded_nan],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(concated_df),columns=concated_df.columns)
```

```python
from sklearn.ensemble import RandomForestRegressor
alg = RandomForestRegressor()
alg.fit(scaled_df,label)
print(alg.score(scaled_df,label)*100)
class Prepare(object):
    '''class docstring'''
    def __init__(self,df)->None:
        self.df = df
    def prepare(self)->None:
        '''function docstring'''
        new_df = self.df.dropna()
        lbl = new_df['quantity']
        new_df = new_df.drop(['quantity','id','date','shop'],axis=1)
        new_df_num = new_df.drop(['city','brand','container','capacity'],axis=1)
        new_df_nan = new_df.drop(new_df_num.columns,axis=1)
        encoded_new_df_nan = pd.get_dummies(new_df_nan)   #Ready
        concated_test_df = pd.concat([new_df_num,encoded_new_df_nan],axis=1)
        scaled_test_df = pd.DataFrame(scaler.fit_transform(concated_test_df),columns=concated_test_df.columns)
        return (scaled_test_df,lbl)
file2 = pd.read_csv('F:\\datasets\\shop_regg\\test.csv')
df_test = file2.copy()
test = Prepare(df_test)
print(alg.score(test.prepare()[0],test.prepare()[1])*100)
```
